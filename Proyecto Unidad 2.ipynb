```python
# Ejercicio 1 y 2
# Celda robusta: detectar encoding/delimiter -> cargar df -> preparar y ejecutar Regresión Lineal Múltiple
import os, io, csv, difflib, warnings
import pandas as pd, numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')

# ------------------- 1) localizar archivo -------------------
candidates = [f for f in os.listdir('.') if f.lower().endswith(('.csv', '.xlsx')) and ('asesin' in f.lower() or 'dataset' in f.lower())]
if not candidates:
    raise FileNotFoundError("No encontré 'DATASET_Asesinatos.csv' en el directorio actual. Sube el archivo o monta Drive.")
fname = candidates[0]
print("Archivo detectado:", fname)

# ------------------- 2) intentar leer CSV probando codificaciones y detectando delimitador -------------------
encodings = ['utf-8', 'utf-8-sig', 'latin1', 'cp1252', 'iso-8859-1', 'utf-16']
delimiters_tried = []
df = None
if fname.lower().endswith('.csv'):
    raw = open(fname, 'rb').read()
    sample_bytes = raw[:20000]  # muestra para sniff
    for enc in encodings:
        try:
            sample_text = sample_bytes.decode(enc)
        except Exception:
            continue
        # intentar sniff del delimitador
        try:
            sniff = csv.Sniffer().sniff(sample_text)
            delim = sniff.delimiter
            delimiters_tried.append((enc, delim))
            try:
                df = pd.read_csv(fname, encoding=enc, sep=delim, engine='python')
                print(f"Lectura correcta con encoding='{enc}' y delimiter='{delim}'")
                break
            except Exception as e:
                # si falló con el delimiter detectado, probar delimitadores comunes
                pass
        except Exception:
            # sniff no funcionó: probar delimitadores comunes
            pass
        # probar delimitadores comunes explícitamente
        for delim_try in [',',';','\t','|']:
            try:
                df = pd.read_csv(fname, encoding=enc, sep=delim_try, engine='python')
                print(f"Lectura correcta con encoding='{enc}' y delimiter='{delim_try}' (prueba explícita)")
                break
            except Exception:
                continue
        if df is not None:
            break
    # si sigue None, intentar lectura laxa con latin1
    if df is None:
        try:
            df = pd.read_csv(fname, encoding='latin1', engine='python')
            print("Lectura fallback con encoding='latin1' sin especificar delimiter.")
        except Exception as e:
            raise UnicodeDecodeError("No pude leer el CSV con las codificaciones probadas. Error final: " + str(e))
else:
    # es Excel
    df = pd.read_excel(fname)
    print("Archivo Excel leído.")

# ------------------- 3) mostrar resumen inicial -------------------
print("\nDimensiones:", df.shape)
print("\nPrimeras filas:")
display(df.head())
print("\nColumnas y tipos:")
print(df.dtypes)

# ------------------- 4) limpiar columnas numéricas que vengan como strings (comas, espacios, miles) -------------------
def to_numeric_clean(s):
    if s.dtype == object:
        s2 = s.astype(str).str.strip().str.replace(r'\s+', '', regex=True)  # quitar espacios
        # si contiene coma y punto, intentar detectar formato: si hay más comas que puntos, asumir coma decimal
        s2 = s2.str.replace(',', '.')  # convertir comas decimales a punto (simple pero efectivo en muchos casos)
        return pd.to_numeric(s2, errors='coerce')
    else:
        return pd.to_numeric(s, errors='coerce')

# aplicarlo de forma cuidadosa solo a columnas que parecen numéricas mezcladas
for c in df.columns:
    # si la columna es object y la mayoría de sus valores parecen números con coma/punto, convertir
    if df[c].dtype == object:
        sample = df[c].dropna().astype(str).head(50).tolist()
        numeric_like = sum(1 for v in sample if any(ch.isdigit() for ch in v)) >= max(1, len(sample)//3)
        if numeric_like:
            converted = to_numeric_clean(df[c])
            # si conversion produjo menos NaNs que mantener strings -> aceptar conversión
            if converted.notna().sum() >= max(1, df[c].notna().sum() * 0.3):
                df[c] = converted

print("\nTipos tras limpieza aproximada:")
print(df.dtypes)

# ------------------- 5) detección automática de target y features (names esperados) -------------------
expected_target_names = [
    'tasa_asesinatos_por_100mil', 'tasa_asesinatos', 'asesinatos_per_100k', 'asesinatos', 'homicidios'
]
expected_features = [
    'indice_gini',
    'tasa_desempleo_juvenil',
    'porcentaje_sin_diploma_secundaria',
    'ingresos_medios',
    'densidad_poblacional',
    'gasto_policial_per_capita',
    'inversion_programas_sociales_per_capita'
]

def match_columns(cols, expected, cutoff=0.6):
    mapping = {}
    for e in expected:
        cl = difflib.get_close_matches(e, cols, n=1, cutoff=cutoff)
        # buscar subcadena
        if not cl:
            for c in cols:
                if e.lower() in c.lower() or c.lower() in e.lower():
                    cl = [c]; break
        mapping[e] = cl[0] if cl else None
    return mapping

cols = df.columns.tolist()
target_map = match_columns(cols, expected_target_names, cutoff=0.6)
detected_target = None
for k,v in target_map.items():
    if v is not None:
        detected_target = v
        break
if detected_target is None:
    # fallback: buscar cualquier columna con 'asesin' o 'homicid'
    for c in cols:
        if 'asesin' in c.lower() or 'homicid' in c.lower():
            detected_target = c
            break

if detected_target is None:
    print("No se pudo detectar automáticamente la columna objetivo. Columnas disponibles:\n", cols)
    raise ValueError("Escribe aquí el nombre de la columna objetivo (por ejemplo 'tasa_asesinatos_por_100mil'), o pega df.columns.tolist().")

print("\nColumna objetivo detectada:", detected_target)

features_map = match_columns(cols, expected_features, cutoff=0.55)
features_present = [features_map[k] for k in expected_features if features_map[k] is not None]
print("\nFeatures mapeadas (esperado -> real):")
for k in expected_features:
    print(f"  {k} -> {features_map.get(k)}")
print("\nFeatures presentes usadas:", features_present)

# ------------------- 6) preparar df_model y limpiar filas NA -------------------
use_cols = [detected_target] + features_present
df_model = df[use_cols].copy()
# asegurarnos numérico
for c in df_model.columns:
    df_model[c] = pd.to_numeric(df_model[c], errors='coerce')
before = df_model.shape[0]
df_model = df_model.dropna()
after = df_model.shape[0]
print(f"\nFilas antes dropna: {before}, después: {after} (eliminadas {before-after})")
if df_model.shape[0] < 10:
    print("Advertencia: quedan menos de 10 filas después de dropna — revisa datos/imputación.")

display(df_model.describe().T)

# ------------------- 7) Regresión Lineal Múltiple (sklearn) -------------------
X = df_model[features_present]
y = df_model[detected_target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
lm = LinearRegression()
lm.fit(X_train, y_train)
y_pred_train = lm.predict(X_train)
y_pred_test  = lm.predict(X_test)

r2_train = r2_score(y_train, y_pred_train)
r2_test  = r2_score(y_test, y_pred_test)
print(f"\nR² (train): {r2_train:.4f}")
print(f"R² (test) : {r2_test:.4f}")
print("Intercept (b0):", lm.intercept_)

coef_df = pd.DataFrame({'feature': features_present, 'coef': lm.coef_}).sort_values(by='coef', key=abs, ascending=False)
display(coef_df)

# ------------------- 8) Interpretación rápida y ejemplo de predicción manual -------------------
for _, row in coef_df.iterrows():
    feat = row['feature']; coef = row['coef']
    signo = "aumenta" if coef > 0 else "disminuye"
    print(f"- Si {feat} aumenta en 1, {detected_target} {signo} en {abs(coef):.4f} (ceteris paribus).")

if X_test.shape[0] > 0:
    x0 = X_test.iloc[0].values
    y_manual = lm.intercept_ + np.dot(lm.coef_, x0)
    print("\nEjemplo predicción manual (1ª fila test):")
    print("X:", X_test.iloc[0].to_dict())
    print("Predicción manual:", y_manual)
    print("Predicción model.predict:", lm.predict([x0])[0])

# ------------------- 9) Gráficos diagnósticos -------------------
plt.figure(figsize=(8,5))
plt.hist(y_test, bins=25, alpha=0.6, label='Real')
plt.hist(y_pred_test, bins=25, alpha=0.6, label='Predicho')
plt.legend(); plt.title('Distribución: reales vs predichos (test)'); plt.xlabel(detected_target); plt.show()

plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred_test, alpha=0.6)
mn, mx = min(y_test.min(), y_pred_test.min()), max(y_test.max(), y_pred_test.max())
plt.plot([mn, mx], [mn, mx], 'r--'); plt.xlabel('Real'); plt.ylabel('Predicho'); plt.title('Real vs Predicho (test)'); plt.show()

resid = y_test - y_pred_test
plt.figure(figsize=(8,4))
plt.scatter(y_pred_test, resid, alpha=0.6)
plt.hlines(0, xmin=y_pred_test.min(), xmax=y_pred_test.max(), linestyles='dashed', colors='r')
plt.xlabel('Predicción (test)'); plt.ylabel('Residuo (real - predicho)'); plt.title('Residual plot (test)'); plt.show()

# ------------------- 10) statsmodels summary y selección por p-value (opcional) -------------------
try:
    import statsmodels.api as sm
    X_train_sm = sm.add_constant(X_train)
    model_sm = sm.OLS(y_train, X_train_sm).fit()
    print("\n--- Resumen statsmodels (OLS) ---")
    print(model_sm.summary())
    sm_coef = pd.DataFrame({'feature': ['const'] + features_present, 'coef': model_sm.params.values, 'pvalue': model_sm.pvalues.values}).sort_values('pvalue')
    display(sm_coef)
    # selección simple por p-value < 0.05
    sig_feats = sm_coef[ (sm_coef['pvalue'] < 0.05) & (sm_coef['feature'] != 'const') ]['feature'].tolist()
    if sig_feats:
        print("Features con p < 0.05:", sig_feats)
        Xs = df_model[sig_feats]; ys = df_model[detected_target]
        Xtr, Xte, ytr, yte = train_test_split(Xs, ys, test_size=0.30, random_state=42)
        lm_s = LinearRegression().fit(Xtr, ytr)
        print("R² subset (train):", r2_score(ytr, lm_s.predict(Xtr)))
        print("R² subset (test) : ", r2_score(yte, lm_s.predict(Xte)))
    else:
        print("No se encontraron features con p < 0.05.")
except Exception as e:
    print("\nNota: statsmodels no está disponible o falló. Continúa con resultados de sklearn. Error:", e)

print("\nProceso completado. Si quieres que adapte nombres exactos de columnas, pega aquí la salida de df.columns.tolist().")
```

```python
# Instalación de scikit-learn si es necesario
# Ejercicio 4

!pip install scikit-learn matplotlib seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# =====================
# Carga de datos
# =====================
# df = pd.read_csv('dataset_regresion_ciudades.csv')
df = pd.read_csv('DATASET_Asesinatos.csv', encoding='latin-1')
variables_iniciales = [
    'indice_gini', 'tasa_desempleo_juvenil', 'porcentaje_sin_diploma_secundaria',
    'ingresos_medios', 'densidad_poblacional',
    'gasto_policial_per_capita', 'inversion_programas_sociales_per_capita'
]
X = df[variables_iniciales]
y = df['tasa_asesinatos_por_100mil']

# División de datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Escalado de variables predictoras
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =====================
# Ejecución 1: SVR Kernel RBF (por defecto)
# =====================
svr_rbf = SVR(kernel='rbf', C=1.0)
svr_rbf.fit(X_train_scaled, y_train)
y_pred_svr_rbf = svr_rbf.predict(X_test_scaled)
r2_svr_rbf = r2_score(y_test, y_pred_svr_rbf)
print(f'R2 SVR (Kernel RBF): {r2_svr_rbf:.4f}')

# =====================
# Ejecución 2: SVR Kernel Linear
# =====================
svr_lin = SVR(kernel='linear', C=1.0)
svr_lin.fit(X_train_scaled, y_train)
y_pred_svr_lin = svr_lin.predict(X_test_scaled)
r2_svr_lin = r2_score(y_test, y_pred_svr_lin)
print(f'R2 SVR (Kernel Linear): {r2_svr_lin:.4f}')

# =====================
# Regresión Lineal Múltiple para comparación
# =====================
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
y_pred_lr = lr.predict(X_test_scaled)
r2_lr = r2_score(y_test, y_pred_lr)
print(f'R2 Regresión Lineal Múltiple: {r2_lr:.4f}')

# =====================
# Comparativa final
# =====================
ejecuciones = pd.DataFrame({
    "Modelo": ['SVR RBF', 'SVR Linear', 'Lineal Múltiple'],
    "R2": [r2_svr_rbf, r2_svr_lin, r2_lr]
})
print(ejecuciones)

# =====================
# GRÁFICO 1: Real vs Predicho (SVR RBF)
# =====================
plt.figure(figsize=(7,6))
sns.scatterplot(x=y_test, y=y_pred_svr_rbf, color='teal', alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title('SVR (Kernel RBF) - Valores Reales vs Predichos')
plt.xlabel('Valores Reales')
plt.ylabel('Valores Predichos')
plt.grid(True)
plt.show()

# =====================
# GRÁFICO 2: Comparativa de R2 entre modelos
# =====================
plt.figure(figsize=(8,6))
sns.barplot(x='Modelo', y='R2', data=ejecuciones, palette='viridis')
plt.title('Comparativa de R² entre Modelos SVR y Lineal')
plt.ylim(0, 1)
plt.ylabel('Coeficiente de Determinación (R²)')
plt.xlabel('Modelo')
for i, r2 in enumerate(ejecuciones['R2']):
    plt.text(i, r2 + 0.01, f'{r2:.3f}', ha='center', fontsize=10)
plt.show()
```

```python
# Ejercicio 3
# Instalación e importación de librerías
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# Carga del dataset
# df = pd.read_csv('dataset_regresion_ciudades.csv')

# Variables predictoras sugeridas
variables_iniciales = [
    'indice_gini', 'tasa_desempleo_juvenil', 'porcentaje_sin_diploma_secundaria',
    'ingresos_medios', 'densidad_poblacional',
    'gasto_policial_per_capita', 'inversion_programas_sociales_per_capita'
]

X = df[variables_iniciales]
y = df['tasa_asesinatos_por_100mil']

# División entrenamiento-prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# ============================
# 1. Ejecución 1: Parámetros por defecto
# ============================
rf1 = RandomForestRegressor(random_state=42)
rf1.fit(X_train, y_train)
y_pred1 = rf1.predict(X_test)
r2_1 = r2_score(y_test, y_pred1)
print(f'R2 (Random Forest por defecto): {r2_1:.4f}')

# Importancia de características
importances1 = pd.Series(rf1.feature_importances_, index=variables_iniciales)
importances1.sort_values(ascending=True).plot(kind='barh', figsize=(8,5), color='teal')
plt.title('Importancia de características - RF (Ejecución 1)')
plt.xlabel('Importancia')
plt.show()

# ============================
# 2. Ejecución 2: Ajuste de Parámetros (más árboles)
# ============================
rf2 = RandomForestRegressor(n_estimators=200, random_state=42)
rf2.fit(X_train, y_train)
y_pred2 = rf2.predict(X_test)
r2_2 = r2_score(y_test, y_pred2)
print(f'R2 (Random Forest con n_estimators=200): {r2_2:.4f}')

importances2 = pd.Series(rf2.feature_importances_, index=variables_iniciales)
importances2.sort_values(ascending=True).plot(kind='barh', figsize=(8,5), color='orange')
plt.title('Importancia de características - RF (Ejecución 2)')
plt.xlabel('Importancia')
plt.show()

# ============================
# 3. Comparativa Final
# ============================
comparativa = pd.DataFrame({
    'Modelo': ['Random Forest - Default', 'Random Forest - 200 Árboles'],
    'R2': [r2_1, r2_2]
})

print('\nResumen Comparativo:')
print(comparativa)

```
